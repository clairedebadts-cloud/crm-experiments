## 1. Load Results

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (14, 6)

df_scored = pd.read_csv('../data/Leads_scored.csv')
print(f"Leads: {len(df_scored):,}")
print(f"Conversion: {df_scored['Converted'].mean()*100:.1f}%")
```

```python
# Segment summary
seg_summary = df_scored.groupby('lead_segment').agg({
    'Converted': ['count', 'mean']
}).round(3)
seg_summary.columns = ['Count', 'Conv_Rate']
seg_summary['Conv_Pct'] = (seg_summary['Conv_Rate'] * 100).round(1)
seg_summary = seg_summary.sort_values('Conv_Rate', ascending=False)

print("Segment Performance:")
print(seg_summary)

cold_count = seg_summary.loc['Cold Lead', 'Count'] if 'Cold Lead' in seg_summary.index else 0
cold_pct = (cold_count / len(df_scored)) * 100
print(f"\n🚨 {cold_count:,} Cold leads ({cold_pct:.1f}%)")
```
